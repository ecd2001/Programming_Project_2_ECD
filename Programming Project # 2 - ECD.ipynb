## The Caesar Cipher

```python
import random

#This function will be used to encrypt a message that the user inputs
def encrypt_msg(message,shift):
    result = ""  #This creates an empty string to store the coded message in
    
    #This for loop goes through each letter in the message, converts it to unicode and shifts it a given number of spaces
    #and then converts the unicode back to letters and stores it in result which it returns to the user
    for i in range(len(message)):
        let = message[i]
        # This if statement will encrypt the uppercase letters
        if (let.isupper()):
            result += chr((ord(let) + shift - 65) % 26 + 65)
        #This else statment will encrypt the lowercase letters
        else:
            result += chr((ord(let) + shift - 97) % 26 + 97)
    return result

#This function will be used to hack an encrypted message
def decrypt_msg(input_message):
    alphabet = "abcdefghijklmnopqrstuvwxyz"
    encrypted_msg = input_message.lower()
    
    for key in range(len(alphabet)):
        translated = ""  #This creates an empty string to store the different letter combinations
        
        # This for loop goes through and prints every shift of the alphabet for the coded message
        for symbol in encrypted_msg:
            if symbol in alphabet:
                num = alphabet.find(symbol)
                num = num - key
                if num < 0:
                    num = num + len(alphabet)
                translated = translated + alphabet[num]
            else:
                translated = translated + symbol
        print(f'Hacking key #{key}: {translated}')

# These prints and inputs welcome the user and start the program
n = input(str("Please enter your name: "))
print("")
print(f"Welcome to the Ceasar Cipher Program {n}!")
print("")
b = input(str("Are you ready to begin the program, yes or no? (no will end the program) "))
print("")

# This while loop allows the user to continue coding or decoding after one time through the program
while b != 'no':
    print(f"Okay {n}, do you want to encrypt a message or hack an encrypted message?")
    x = int(input("Enter 1 to encrypt a message. Enter 2 to hack an encrypted message. (0 to stop): "))
    if x == 1:
        message = input(str("Please enter the message you want encrypted: "))
        #I gave the user the option to choose how many places they want to shift their message, if they say no
        #then the program produces a random integer for the shift
        q = input(str("Do you want to choose how many places your message is shifted? (yes or no): "))
        if q == "yes":
            shift = int(input("Please enter the number of places you want to shift by between 1 and 26: "))
        else:
            shift = int(random.randrange(1,27))
        print(f'Your encrypted message is: {encrypt_msg(message,shift)}')
        print("")
        #This input allows the user to go through the program agian or end it
        c = input(str("Do you want to use the program again, yes or no? "))
        if c == "no":
            print("Goodbye!")
            break
        else:
            continue
    elif x == 2:
        input_message = input(str("Please enter the encrypted message that you want to hack: "))
        print("")
        print("IMPORTANT! When looking for the translated message, note that spaces will be 'n' and symbols will be random letters!")
        print("")
        decrypt_msg(input_message)
        print("")
        #This input allows the user to go through the program agian or end it
        c = input(str("Do you want to use the program again, yes or no? "))
        if c == "no":
            print("Goodbye!")
            break
        else:
            continue
    else:
        print("Goodbye!")
        break
```
